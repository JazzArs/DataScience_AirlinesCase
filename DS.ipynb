```python
# Convert dates in datetime
df['flight_date_dt'] = pd.to_datetime(df['flight_date'], dayfirst=True, errors='coerce')


# Sort by ID and date
# df = df.sort_values(['ID', 'flight_date_dt'])

sus_ids_30d = []

for pid, group in df.groupby('ID'):
    dates = group['flight_date_dt'].values
    # using sliding window
    for i in range(len(dates)):
        window_start = dates[i]
        window_end = window_start + pd.Timedelta(days=30)
        count = ((dates >= window_start) & (dates < window_end)).sum()
        if count > 5:
            sus_ids_30d.append(pid)
            break

sus_ids_30d = list(set(sus_ids_30d))
print(len(sus_ids_30d))
print(sus_ids_30d[:30])

with open('sus_ids_30d.txt', 'w', encoding='utf-8') as f:
    for sus in sus_ids_30d:
        f.write(sus + '\n')
```

Создадим булевую маску, которая определяет междугородние перелеты.

```python
df['is_Russia'] = df['arr_city'].isin(Russian_cities['city'])
df = df.reset_index(drop=True)
display(df['is_Russia'])
print(df.columns)
df.head()
```

Лица, совершившие въезд на территорию Российской Федерации с последующими вылетами в третьи страны, вызывают подозрения.

```python
sus_ids_russia_intl = []

for pid, group in df.groupby('ID'):
    # Ensure the group is sorted by date for the sliding window
    group = group.sort_values('flight_date_dt')
    dates = group['flight_date_dt'].values
    is_russia_values = group['is_Russia'].values

    for i in range(len(dates)):
        window_start = dates[i]
        window_end = window_start + pd.Timedelta(days=30)

        # Get the indices within the 30-day window
        window_indices = (dates >= window_start) & (dates < window_end)

        # Count Russian and international flights within the window
        russian_flights_count = is_russia_values[window_indices].sum()
        international_flights_count = (~is_russia_values[window_indices]).sum()

        # Check if the conditions are met (at least 1 Russian and at least 2 international)
        if russian_flights_count >= 1 and international_flights_count >= 2:
            sus_ids_russia_intl.append(pid)
            break  # Add the ID and move to the next ID

# Get unique suspicious IDs
sus_ids_russia_intl = list(set(sus_ids_russia_intl))

print(len(sus_ids_russia_intl))
print(sus_ids_russia_intl[:30])

with open('sus_ids_russia_intl.txt', 'w', encoding='utf-8') as f:
    for sus in sus_ids_russia_intl:
        f.write(sus + '\n')
```

Подведем итоги, приведя список лиц не вызываюзих доверия, которые попадают под несколько критериев.

```python
# Combine all suspicious ID lists with assigned weights based on significance
# Weights: sus_ticket (8), sus_russia_intl (2), sus_30d (4), sus_docs (1)
all_sus_ids_weighted = {
    'sus_ticket': (sus_ids_t, 8),
    'sus_russia_intl': (sus_ids_russia_intl, 2),
    'sus_30d': (sus_ids_30d, 4),
    'sus_docs': (sus_ids_d, 1)
}

# Create a dictionary to store the weighted score for each ID
id_weighted_score = {}

# Iterate through each list of suspicious IDs and their weight
for criteria, (ids, weight) in all_sus_ids_weighted.items():
    for id_val in ids:
        if id_val in id_weighted_score:
            id_weighted_score[id_val] += weight
        else:
            id_weighted_score[id_val] = weight

# Convert the dictionary to a list of tuples and sort by the weighted score in descending order
prioritized_sus_ids_weighted = sorted(id_weighted_score.items(), key=lambda item: item[1], reverse=True)

print("Prioritized Suspicious IDs (ID, Weighted Score):")
for id_val, score in prioritized_sus_ids_weighted[:30]: # Print the top 30
    print(f"{id_val}: {score}")

# Count the number of individuals with a weighted score of 10
count_score_15 = sum(1 for id_val, score in prioritized_sus_ids_weighted if score == 15)
print(f"\nNumber of individuals with a weighted score of 15: {count_score_15}")

# Save all prioritized suspicious IDs and scores to a text file
with open('prioritized_sus_ids_weighted.txt', 'w', encoding='utf-8') as f:
    for id_val, score in prioritized_sus_ids_weighted:
        f.write(f"{id_val}: {score}\n")

# Save IDs with a weighted score of 15 to a separate text file
with open('prioritized_sus_ids_weighted_top.txt', 'w', encoding='utf-8') as f:
    for id_val, score in prioritized_sus_ids_weighted:
        if score == 15:
            f.write(f"{id_val}\n")
```
